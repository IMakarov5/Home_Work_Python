### Условие 3: 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
import numpy as np

corr_matrix = df[['price',
                  'bedrooms', 
                  'bathrooms',	
                  'sqft_living',	
                  'sqft_lot',	
                  'floors',	
                  'waterfront',	
                  'view',
                  'condition',
                  'grade',
                  'sqft_above',
                  'sqft_basement',
                  'yr_built',
                  'yr_renovated']].corr()
                        
                        
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
import seaborn as sns

plt.figure(figsize=(6, 4))
plt.title('Матрицу корреляций')
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

Вывод: на стоимость не влияет:
             sqft_lot - Кв. метры общие
             condition - Значение от 1 до 5 насколько хорошее состояние
             yr_built - Год постройки дома
             yr_renovated - Год ремонта дома

```python
# Готовим данные для графика в виде сводной таблицы

data = df.groupby('grade').agg({'price': 'mean'}).reset_index().sort_values(by='grade')

data
```

```python
# Подготовка переменных
names = data['grade']
values = data['price']

# Bar
plt.figure(figsize=(8, 6))
plt.bar(names, values)

# Опесание графика
plt.title('Распределение стоимости дома от конструкция здания и дизайн')
plt.xlabel('Значения конструкции здания и дизайн')
plt.ylabel('Стоимость')
plt.xticks(rotation=50);
```

Вывод: чем выше значение конструкции здания и зизайна, тем выше стоимость
