## 1. Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800]. Найдите ее среднее значение и дисперсию.


```python
a=200
b=800

M = (a+b)/2

D= ((b-a)**2)/12

print(f'Cреднее значение {M}')
print(f'Дисперсия {D}')
```

```python
D = 0.2
a = 0.5

b = ((12 * D)**0.5) + a

M = (a+b)/2


print(f'Правая граница величины B {b}')
print(f'Cреднее значение равномерно распределенной величины B: {M}')
```
