## Задание 1
Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
salary = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]


salary_avg = sum ([ value for value in salary ]) / len(salary)

print ('Среднее арифметическое:', salary_avg)




S2 = sum ([ (value - salary_avg) ** 2 for value in salary ]) / len(salary)

print ('Cмещенная оценка дисперсий для данной выборки: ', S2)


S2 = sum ([ (value - salary_avg) ** 2 for value in salary ]) / (len(salary) - 1)

print('Несмещенная оценка дисперсий для данной выборки: ', S2)

sig = S2 ** 0.5

print ('Среднее квадратичное отклонение: ', sig)


```

## Задание 2
В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

A11 - из первого ящика вынули 2 белых мяча С(5,2)/С(8,2)

А12 - из второго ящика вынули 1 белый и 3 черных С(5,1)*C(7,3)/C(12,4)

А21 - из первого ящика вынули 2 черных C(3,2)/C(8,2)

А22 - из вторго ящика вынули 3 белых и 1 черный C(5,3)*C(7,1)/C(12,4)

А31 - из первого ящика вынули 1 белый  и 1 черный C(5,1)*C(3,1)/C(8,2)

А32 - из второго ящика вынули 2 белый и 2 черных C(5,2)*C(7,2)/C(12,4)

```python
A11 = C(5,2) / C(8,2)
A12 = (C(5,1) * C((12-5),3)) / C(12,4)

A21 = C((8-5),2) / C(8,2)
A22 = (C(5,3) * C((12-5),1)) / C(12,4)

A31 = (C(5,1) * C((8-5),1)) / C(8,2)
A32 = (C(5,2) * C((12-5),2)) / C(12,4)



p = A11 * A12 + A21 * A22 + A31 * A32
    

print('Вероятность того, что ровно три мяча белые: ', p )
```

## Задание 3
На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: 
* a). первым спортсменом 
* б). вторым спортсменом 
* в). третьим спортсменом.

```python
P = 1/3 * 0.9 + 1/3 * 0.8 + 1/3 * 0.6 
print ('Вероятность того, что мишень будет поражена: ',P)

print('Вероятность того, что выстрел произведен первым спортсменом: ', (1/3 * 0.9) * P)
print('Вероятность того, что выстрел произведен вторым спортсменом: ', (1/3 * 0.8) * P)
print('Вероятность того, что выстрел произведен третьим спортсменом: ', (1/3 * 0.6) * P)
```

## Задание 5
Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя: а). все детали б). только две детали в). хотя бы одна деталь г). от одной до двух деталей?

```python
p1 = 0.1
p2 = 0.2
p3 = 0.25


Pa = p1*p2*p3

print('Вероятность того, что в первый месяц выйдут из строя: все детали ', Pa)

Pb = p1*p2*(1-p3)+p1*p3*(1-p2)+p2*p3*(1-p1)

print('Вероятность того, что в первый месяц выйдут из строя: только две детали ', Pb)


Pv = 1-(1-p1)*(1-p2)*(1-p3)

print('Вероятность того, что в первый месяц выйдут из строя: хотя бы одна деталь ', Pv)



Pg = p1*(1-p2)*(1-p3)+p2*(1-p1)*(1-p3)+p3*(1-p1)*(1-p2)

print('Вероятность того, что в первый месяц выйдут из строя: от одной до двух деталей ', Pg+Pv)

```
