1. Из колоды в 52 карты извлекаются случайным образом 4 карты. 
* Найти вероятность того, что все карты – крести. 
* Найти вероятность, что среди 4-х карт окажется хотя бы один туз.

Вероятность события — это отношение числа элементарных исходов, благоприятствующих данному событию, к числу всех равновозможных исходов опыта, в котором оно может появиться. 

P(А) = m/n

m - число элементарных исходов, благоприятствующих данному событию

n - число всех равновозможных исходов опыта, в котором оно может появиться

Сочетания – это набор, состоящий из k  элементов, выбранных  из множества, содержащего n  различных элементов.

Число элементарных исходов, благоприятствующих данному событию:

С(n,k) = n! / k! (n-k)!

* Найти вероятность того, что все карты – крести:

для подсчета m:

k - число карт 4

n - число крести в колоде 52/4 = 13

С(13,4)

для подсчета n:

k - число карт 4

n - число всех карт, то есть 52

С(52,4)

```python
from math import factorial

def C (n: int, k: int) -> float:
    return (factorial(n) / (factorial(k) * factorial(n - k))) 

def A (n: int, k: int) -> float:
    return (factorial(n) / factorial(n - k)) 

def Pn (n: int, ) -> float:
    return factorial(n)
```

```python
m = C (13,4)
print(m)

n = C (54,4)
print(n)

P=m/n
print ('Вероятность того, что все карты – крести', P) 
```
