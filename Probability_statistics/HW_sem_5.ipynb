Нуливая гипотиза:
H_0

m > m0

n = 100**0.5 = 10

z = (17.5 - 17)/(2/10) = 2.5

2.5 - расчетный Z критерий по выборке

zt = 1.645
 
Расчетный Z критерий больше, чем табличное Z критерий. Получается, что расчетное значение 2.5 НЕ попадает в область принятия гипотезы H_0.

## Задачу 4 решать с помощью функции. 4.Есть ли статистически значимые различия в росте дочерей? Рост матерей 172, 177, 158, 170, 178,175, 164, 160, 169, 165 Рост взрослых дочерей: 173, 175, 162, 174, 175, 168, 155, 170, 160

```python
x = [172, 177, 158, 170, 178,175, 164, 160, 169, 165]
y = [173, 175, 162, 174, 175, 168, 155, 170, 160]  

x_avg = sum ([ value for value in x]) / len(x)
y_avg = sum ([ value for value in y]) / len(y)

print (f'Среднее значение росто матерей {x_avg}')
print (f'Среднее значение росто дочерей {y_avg}')

nx = len(x)
ny = len(y)

sd_x = (sum ([ (x_avg-value)**2 for value in x]) / nx)**0.5
sd_y = (sum ([ (y_avg-value)**2 for value in y]) / ny)**0.5

print (f'Dx:  {sd_x}')
print (f'Dy: {sd_y}')


t = (x_avg - y_avg)/ ((sd_x**2/nx)+ (sd_y**2/ny))**0.5

print (f't:  {t}')






```
