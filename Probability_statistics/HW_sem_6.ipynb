## 1 Известно, что генеральная совокупность распределена нормально со средним квадратическим отклонением, равным 16. Найти доверительный интервал для оценки математического ожидания a с надежностью 0.95, если выборочная средняя M = 80, а объем выборки n = 256

```python
lam = 0.95
tlam = 1.96

S = 16

M = 80

n = 256

D = tlam * S / (n ** 0.5)

a = M - D
b = M + D

print(f'Доверительный интервал: [{a}, {b}]')

```

## 2.В результате 10 независимых измерений некоторой величины X, выполненных с одинаковой точностью, получены опытные данные: 6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1 Предполагая, что результаты измерений подчинены нормальному закону распределения вероятностей, оценить истинное значение величины X при помощи доверительного интервала, покрывающего это значение с доверительной вероятностью 0,95. 

```python
lam = 0.95
tlam = 2.26

x = [6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1]

M = sum ([ value for value in x]) / len(x)
print(f'Среднее арифметическое для выборки: {M}')



S = sum ([ (value - M) ** 2 for value in x ]) / (len(x) - 1)
sig = S ** 0.5
print (f'Среднее квадратичное отклонение: {sig}')


n = 10

D = tlam * sig / (n ** 0.5)

a = M - D
b = M + D

print(f'Доверительный интервал: [{a}, {b}]')


```

## 3.Рост дочерей 175, 167, 154, 174, 178, 148, 160, 167, 169, 170 Рост матерей  178, 165, 165, 173, 168, 155, 160, 164, 178, 175 Используя эти данные построить 95% доверительный интервал для разности среднего роста родителей и детей.

```python
import numpy as np
import scipy.stats as stats


lam = 0.95


x = np.array([178, 165, 165, 173, 168, 155, 160, 164, 178, 175])
y = np.array([175, 167, 154, 174, 178, 148, 160, 167, 169, 170])

# среднее арифмитическое для выборки
x_avg = np.mean(x)
y_avg = np.mean(y)

# разность средних
delta = x_avg - y_avg

# несмещенную дисперсию для выборки
Dx = np.var(x, ddof=1)
Dy = np.var(y, ddof=1)

# объединенная оценка дисперсии
D= (Dx+Dy)/2

print(f'Объединенная оценка дисперсии: {D}')

# стандартная ошибка разности среднего

SE= (D/10+D/10)**0.5
print(f'Стандартная ошибка разности среднего: {SE}')


# критерий Стьюдента
ts = stats.t.ppf(0.95, (len(x) + len(y) - 2))

print(f'Критерий Стьюдента: {ts}')

# нижняя граница интеравала

a = delta - ts*SE

# верхняя граница интеравала

b = delta + ts*SE

print(f'Доверительный интервал: [{a}, {b}]')
```
